```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_path = "C:/Users/marcc/Downloads/aiperia/00_coding_challenge/00_coding_challenge/data"

# Load data from all customers
df_1001 = pd.read_parquet(f"{data_path}/customer_1001/03_primary/app_dataset.parquet")
df_1002 = pd.read_parquet(f"{data_path}/customer_1002/03_primary/app_dataset.parquet")
df_1003 = pd.read_parquet(f"{data_path}/customer_1003/03_primary/app_dataset.parquet")

df = pd.concat([df_1001, df_1002, df_1003], ignore_index=True)

print(f"Loaded {len(df)} records from 3 customers")
```

```python
id_store = 100190001
id_product = 10010001

plotting_data = df.query('id_store == @id_store & id_product == @id_product')
plotting_data = plotting_data.sort_values("target_date")

plt.figure(figsize=(12, 6))

# Plot delivery_qty as a line
plt.plot(plotting_data["target_date"], plotting_data["delivery_qty"], color="orange", label="Delivery Qty", linewidth=2)

# Plot sales_qty as a bright blue bar
plt.bar(plotting_data["target_date"], plotting_data["sales_qty"], color="#4FC3F7", label="Sales Qty")

# Plot return_qty as a dark blue bar stacked on top of sales_qty
plt.bar(plotting_data["target_date"], plotting_data["return_qty"], bottom=plotting_data["sales_qty"], color="#1565C0", label="Return Qty")

plt.xlabel("Date")
plt.ylabel("Quantity")
plt.title("Daily Delivery, Sales, and Returns")
plt.legend()
plt.tight_layout()
plt.show()
```
